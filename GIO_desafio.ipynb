```python
import pandas as pd
import json

```

```python
sdw2023_api_url = 'https://sdw-2023-prd.up.railway.app'
```

```python
users = [
    {'id': 1, 'nome': 'Naruto', 'news': []},
    {'id': 2, 'nome': 'Hinata', 'news': []}
]
```

```python
print("usuario carregados: ")
users
```

```python
def generate_ai_news(user):
    return f"{user['nome']}, investir é essencial para construir um futuro financeiro sólido."

for user in users:
    news = generate_ai_news(user)
    user['news'].append({
        "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
        "description": news
    })

print("Mensagens geradas:")
users
```

```python
with open("users_with_news.json", "w", encoding="utf-8") as f:
    json.dump(users, f, ensure_ascii=False, indent=2)

print("Arquivo JSON salvo com sucesso!")
```
